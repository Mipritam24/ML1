# Backward Elimination Method

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

### Step 1 : Read the dataset

```python
import pandas as pd 
df = pd.read_csv("Cars93.csv", na_values=["","NA"], keep_default_na= False)
df.head()
```

```python
df.info()
```

### Step 3 Separate X(variables) and Y(label)

```python
X= df.drop(columns=["id","Weight"])
Y= df[["Weight"]]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4 : Preprocessing of the data by feature selection pipeline

```python
cat = list(X.columns[X.dtypes=="object"])
con = list (X.columns[X.dtypes !="object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe= make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder()
)
```

```python
pre1= ColumnTransformer([
    ("num",num_pipe, con),
    ("cat", cat_pipe,cat)
]).set_output(transform="pandas")
```

```python
pre1
```

```python
X_pre1= pre1.fit_transform(X)
X_pre1
```

### Step 5 : Feature Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sel = SequentialFeatureSelector(LinearRegression(), direction="backward", n_features_to_select="auto")
sel.fit(X_pre1, Y)
imp_cols = sel.get_feature_names_out()
imp_cols
```

```python
sel_cols = [col.split("__")[1] for col in imp_cols]
sel_cols
```

```python
X_sel= X[sel_cols]
X_sel
```

### Step 6 : Create final pipeline

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=="object"])
con_sel = list(X_sel.columns[X_sel.dtypes!="object"])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe2 = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop= "first")
)
```

```python
pre2= ColumnTransformer([
    ("num", num_pipe2, con_sel),
    ("cat",cat_pipe2, cat_sel)]
).set_output(transform="pandas")
```

```python
pre2
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre
```

### Step 7: Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre, Y, test_size= 0.2, random_state= 21)
```

### Step 8 Model Building

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Step 9 : Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
def model_evaluation(model, X, Y):
    ypred = model.predict(X)
    mse = mean_squared_error(Y, ypred)
    rmse= mse**(1/2)
    mae= mean_absolute_error(Y, ypred)
    mape= mean_absolute_percentage_error(Y, ypred)
    r2= r2_score(Y,ypred)
    print(f"MSE: {mse:.2f}")
    print(f"RMSE:{rmse:.2f}")
    print(f"MAE:{mae:.2f}")
    print(f"MAPE:{mape:.2f}")
    print(f"R2_score:{r2:.4f}")
```

```python
model_evaluation(model, xtrain, ytrain)
```

```python
model_evaluation(model, xtest, ytest)
```
